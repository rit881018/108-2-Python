```python
sen=df["Content"]
print(sen[0])
sen=list(sen)
```

```python
labels= (df["Sentimen"])
print(labels[0])
print(type(labels[0]))
labels=list(labels)
```

```python
for i in range(len(sen)):
    sen[i]=jieba.lcut(str(sen[i]))
print(sen[0])
sen1=sen.copy()
```

```python
from collections import Counter
words = Counter()  # Dictionary that will map a word to the number of times it appeared in all the training sentences
for i, sentence in enumerate(sen):
    for word in sen[i]:
        words.update([word.lower()])
```

```python
words = {k:v for k,v in words.items() if v>1}
print(words)
# Sorting the words according to the number of appearances, with the most common word being first
words = sorted(words, key=words.get, reverse=True)
```

```python
print(len(words))
print(words)
```

```python
#words = ['_PAD','_UNK'] + words
# Dictionaries to store the word to index mappings and vice versa
word2idx = {o:i for i,o in enumerate(words)}
idx2word = {i:o for i,o in enumerate(words)}
```

```python
print(word2idx['的'])
print(idx2word[1000])
```

```python
for i, s in enumerate(sen):
    # Looking up the mapping dictionary and assigning the index to the respective words
    sen[i] = [word2idx[word] if word in word2idx else 0 for word in s]
```

```python
print(len(sen[0]))
print(sen[11002])
print(sen1[11002])
```

```python
import numpy as np
# Defining a function that either shortens sentences or pads sentences with 0 to a fixed length
def pad_input(sentences, seq_len):
    features = np.zeros((len(sentences), seq_len),dtype=int)
    for ii, review in enumerate(sentences):
        if len(review) != 0:
            features[ii, -len(review):] = np.array(review)[:seq_len]
    return features

seq_len = 30  # The length that the sentences will be padded/shortened to

train_sentences, val_sentences=sen[:9000], sen[9000:];
train_labels, val_labels=labels[:9000], labels[9000:]

train_sentences = pad_input(train_sentences, seq_len)
val_sentences = pad_input(val_sentences, seq_len)

# Converting our labels into numpy arrays
train_labels = np.array(train_labels)
val_labels = np.array(val_labels)
```

```python
print(train_sentences[0])
print(val_sentences[2004])
print(train_sentences.shape)
print(val_sentences.shape)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers.embeddings import Embedding
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
train_labels=keras.utils.to_categorical(train_labels, 2)
val_labels=keras.utils.to_categorical(val_labels, 2)
print(train_labels.shape)
print(val_labels.shape)
```

```python
train_sentences = np.array(train_sentences)
val_sentences = np.array(val_sentences)
print(train_sentences.shape)
re_train_sentences=np.reshape(train_sentences, (train_sentences.shape[0], train_sentences.shape[1],1))
re_val_sentences=np.reshape(val_sentences, (val_sentences.shape[0],val_sentences.shape[1],1))
#re_train_sentences=train_sentences[:,:,np.newaxis]
print(re_train_sentences.shape)
```

```python
top_words=len(words)

model = Sequential()
# Adding the first LSTM layer and some Dropout regularisation
#model.add(Embedding(input_dim=top_words+1, output_dim=seq_len))
model.add(LSTM(units=100, return_sequences=True,  input_shape=(train_sentences.shape[1],1)))
model.add(Dropout(0.1))
model.add(LSTM(units=50, return_sequences=True,))
model.add(Dropout(0.1))
model.add(LSTM(50))
model.add(Dropout(0.1))
model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
train_history = model.fit(x=re_train_sentences, y=train_labels, validation_split=0.1, epochs=30, batch_size=512) 
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

history_dict = train_history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']

epochs = range(1, len(history_dict['accuracy']) + 1)

plt.plot(train_history.history['accuracy'])
plt.plot(train_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(epochs, loss_values, 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
scores = model.evaluate(re_val_sentences, val_labels)  
```
